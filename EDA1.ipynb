```python
Q3=df[['LB','AC','DS','DP','MLTV','Width','Tendency','NSP']].quantile(0.75)
```

```python
IQR=Q3-Q1
```

```python
IQR
```

```python
UL=Q1+1.5*IQR
LL=Q3-1.5*IQR
```

```python
columns_of_interest = ['LB', 'AC', 'DS', 'DP', 'MLTV', 'Width', 'Tendency', 'NSP']
limits_df = pd.DataFrame({
    'Column': columns_of_interest,
    'UL': UL.values,
    'LL': LL.values
})

# Combine UL and LL into a single tuple column
limits_df['Formatted'] = list(zip(limits_df['UL'], limits_df['LL']))

# Display the result
print(limits_df[['Column', 'Formatted']])
```

```python
columns_of_interest = ['LB', 'AC', 'DS', 'DP', 'MLTV', 'Width', 'Tendency', 'NSP']
mask = pd.DataFrame()
for column in columns_of_interest:
    mask[column] = (df_imp[column] < LL[column]) | (df_imp[column] > UL[column])
    outliers = df_imp[mask.any(axis=1)]

print(outliers)
```

```python
# for HISTOGRAMS
# Define columns of interest
columns_of_interest = ['LB', 'AC', 'DS', 'DP', 'MLTV', 'Width', 'Tendency', 'NSP']

# Plot the distribution for each column
for column in columns_of_interest:
    #plt.figure(figsize=(10, 6))
    sns.histplot(df_imp[column], kde=True,color='black')  # `histplot` is the updated function replacing `distplot`
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#for Scatter plots
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plots for pairs of columns
for i, column1 in enumerate(columns_of_interest):
    for column2 in columns_of_interest[i+1:]:
        plt.figure(figsize=(10, 6))
        plt.scatter(df_imp[column1], df_imp[column2], alpha=0.5,color='orange')
        plt.xlabel(column1)
        plt.ylabel(column2)
        plt.title(f'Scatter Plot of {column1} vs {column2}')
        plt.show()
```

```python
INSIGHTS:
1. Understanding Fetal Health
Baseline Fetal Heart Rate (LB):

The average baseline fetal heart rate (LB) is an important indicator of fetal well-being. A normal range typically lies between 110 to 160 beats per minute. Our dataset's analysis showed most values within this range, indicating a generally healthy fetal population.
Recommendation: Continuous monitoring and regular check-ups for pregnant women to keep track of the baseline fetal heart rate.
Fetal Movement (FM):

Higher fetal movements (FM) are generally a sign of a healthy fetus. The analysis showed a positive correlation between fetal movements and other positive fetal health indicators.
Recommendation: Advise pregnant women to be aware of fetal movements and report any significant changes to healthcare providers.
2. Anomalies and Outliers
Outliers in Baseline Heart Rate:

Outliers in LB might indicate abnormal fetal conditions, such as hypoxia or other stress conditions.
Recommendation: Any significant deviations from the normal range should prompt immediate medical evaluation to prevent complications.
Fetal Distress Indicators:

Variables like 'ASTV' (percentage of time with abnormal short-term variability) and 'ALTV' (percentage of time with abnormal long-term variability) can indicate potential fetal distress.
Recommendation: Increased monitoring and possibly immediate intervention if high values in these parameters are observed.
3. Data Distribution and Trends
Distribution of Fetal Health Status (NSP):
The 'NSP' variable categorizes fetal health status into Normal, Suspect, and Pathological. The majority of the data points fall into the Normal category, but the presence of Suspect and Pathological cases indicates a need for vigilant monitoring.
Recommendation: Establish protocols for more frequent and detailed monitoring of cases falling into the Suspect and Pathological categories.
4. Advanced Visualizations
Violin Plots for Health Status:
Violin plots showed the distribution of heart rate variability and movements across different health statuses. Pathological cases typically had lower baseline variability and movement.
Recommendation: Use such visual tools in clinical settings to quickly identify high-risk pregnancies.
```
